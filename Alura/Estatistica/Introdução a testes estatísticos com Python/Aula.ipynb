### Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set()
```

```python
tmdb = pd.read_csv('dados/tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(tmdb['vote_average'])
ax.set_title('Histograma das notas dos filmes', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Densidade', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(tmdb['vote_average'], kde=False)
ax.set_title('Histograma das notas dos filmes', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(8,6))
ax = sns.boxplot(x=tmdb['vote_average'], width=0.5)
ax.set_title('Distribuição de notas dos filmes do TMDB 5000', fontsize=18)
ax.set_xlabel('Nota média do filme', fontsize=14)
```

```python
print('Número de votos que os filmes com nota 0 tiveram:')
tmdb.query('vote_average == 0')['vote_count'].value_counts()
```

É possível perceber que alguns filmes tiveram poucos votos (ou até mesmo nenhum). Decidimos que esses filmes não devem pertencer a nossa análise. 

```python
tmdb_com_mais_de_10_votos = tmdb.query('vote_count >= 10')
tmdb_com_mais_de_10_votos.describe()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(tmdb_com_mais_de_10_votos['vote_average'], kde=False)
ax.set_title('Histograma das notas médias dos filmes com pelo menos 10 votos', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)

media_das_notas_dos_filmes = tmdb_com_mais_de_10_votos['vote_average'].mean()
ax.axvline(media_das_notas_dos_filmes, color='k', linestyle='--')
ax.annotate('Média', xy=(media_das_notas_dos_filmes+0.4, 400), fontsize=18)

plt.show()
```

### Analisando os dados do movie lens

```python
notas = pd.read_csv('dados/ratings.csv')
notas.head()
```

```python
notas.shape
```

```python
media_das_notas_movie_lens = notas.groupby('movieId').mean()['rating']
media_das_notas_movie_lens.head()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(media_das_notas_movie_lens, kde=False)
ax.set_title('Histograma das notas dos filmes', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Frequência', fontsize=14)

plt.show()
```

```python
filmes_com_pelo_menos_10_votos = notas.groupby('movieId').count().query('rating >=10').index
filmes_com_pelo_menos_10_votos.values
```

```python
nota_media_dos_filmes_com_pelo_menos_10_votos = media_das_notas_movie_lens.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set_title('Histograma das notas dos filmes no Movie Lens com pelo menos 10', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Distribuição', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set_title('Distribuição das notas médias dos filmes no Movie Lens com pelo menos 10 votos', fontsize=18)
ax.set_xlabel('Nota média dos filmes', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos, 
                  hist_kws={'cumulative':True}, 
                  kde_kws={'cumulative':True})
ax.set_title('Histograma das notas dos filmes no Movie Lens com pelo menos 10 votos', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Proporção acumulada de filmes', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.distplot(tmdb_com_mais_de_10_votos['vote_average'], 
                  hist_kws={'cumulative':True}, 
                  kde_kws={'cumulative':True})
ax.set_title('Histograma das notas dos filmes no Movie Lens com pelo menos 10 votos', fontsize=18)
ax.set_xlabel('Média dos votos', fontsize=14)
ax.set_ylabel('Proporção acumulada de filmes', fontsize=14)

plt.show()
```
