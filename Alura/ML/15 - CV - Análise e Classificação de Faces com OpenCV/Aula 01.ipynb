## Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import dlib
```

## Importando os dados

```python
imagem = cv2.imread('imagens/imagem.jpg')
ax = plt.imshow(imagem)
plt.axis(False)
plt.show()
```

```python
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
ax = plt.imshow(imagem_rgb)
plt.axis(False)
plt.show()
```

```python
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
ax = plt.imshow(imagem_gray, cmap='gray')
plt.axis(False)
plt.show()
```

```python
imagem = cv2.imread('imagens/px-people.jpg')
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
ax = plt.imshow(imagem)
plt.show()
```

```python
imagem.shape
```

```python
imagem_roi = imagem[100:200,1000:1200]
ax = plt.imshow(imagem_roi)
plt.show()
```

```python
imagem_roi.shape
```

```python
cv2.imwrite('output/imagem_roi.png', imagem_roi)
```

```python
imagem_roi_bgr = cv2.cvtColor(imagem_roi, cv2.COLOR_RGB2BGR)
cv2.imwrite('output/imagem_roi.png', imagem_roi_bgr)
```

```python
imagem_crianca = cv2.imread('imagens/crianca.png')
imagem_crianca = cv2.cvtColor(imagem_crianca, cv2.COLOR_BGR2RGB)
ax = plt.imshow(imagem_crianca)
plt.show()
```

```python
crianca_roi = imagem_crianca[200:400, 520:680]
crianca_roi_gray = cv2.cvtColor(crianca_roi, cv2.COLOR_RGB2GRAY)
crianca_roi_gray = cv2.cvtColor(crianca_roi_gray, cv2.COLOR_GRAY2RGB)
imagem_crianca_destacada = imagem_crianca.copy()
imagem_crianca_destacada[200:400, 520:680] = crianca_roi_gray

ax = plt.imshow(imagem_crianca_destacada)
plt.show()
```

```python
imagem_crianca_destacada = cv2.cvtColor(imagem_crianca_destacada, cv2.COLOR_RGB2BGR)
cv2.imwrite('output/imagem_crianca_destacada.jpg', imagem_crianca_destacada)
```
