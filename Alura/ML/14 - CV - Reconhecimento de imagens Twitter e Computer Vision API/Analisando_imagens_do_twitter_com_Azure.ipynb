## Streaming + vision API

```python
famosos_id = []
for famoso in famosos:
  try:
    famosos_id.append(api.get_user(famoso).id_str)
  except tw.TweepError as e:
    print(famoso)
    continue
```

```python
class MinhaStreamListener(tw.StreamListener):
  def on_status(self, status):
    print('Usu√°rio: ', status.user.screen_name)
    print('texto: ', status.text)

    if 'media' in status.entities:
      url = status.entities['media'][0]['media_url']
      print('URL: ', url)

      analise_celebridade = client.analyze_image_by_domain('celebrities', url, 'en')
      lista_celebridades = [celebridade['name'] for celebridade in analise_celebridade.result['celebrities']]
      print(lista_celebridades)

      descricao = client.describe_image(url, 1, 'en')
      texto_descricacao = descricao.captions[0].text
      print('Descri√ß√£o: ', texto_descricacao)

      analise_imagem = client.analyze_image(url, [VisualFeatureTypes.tags])
      lista_tags = [tag.name for tag in analise_imagem.tags]
      print('Tag: ', lista_tags)

      resultados = {'usuario': status.user.screen_name,
                    'texto': status.text,
                    'url': url,
                    'celebridades': lista_celebridades,
                    'descricacao': texto_descricacao,
                    'tags': lista_tags}

      with open('tweets.txt', 'a') as f:
        f.write(json.dumps(resultados))
        f.write('\n')

    print('-----------\n')
```

```python
minhaStream = None
minhaStream = tw.Stream(auth=auth, listener=MinhaStreamListener())
minhaStream.filter(follow=famosos_id)
```
