```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='g')
ax.set_title('Temperatura no tempo', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='best', fontsize=14)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])
x2 = df['data'][df['temperatura'].idxmax() - 7000]
y2 = max(df['temperatura']) - 5

ax.annotate('Máximo', xy=(x1,y1), fontsize=14, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

x1 = df.loc[df['temperatura'].idxmin(), 'data']
y1 = min(df['temperatura'])
x2 = df.loc[df['temperatura'].idxmin() + 5000, 'data']
y2 = min(df['temperatura']) + 5

ax.axhline(max(df['temperatura']), color='k', ls='--')
ax.axhline(min(df['temperatura']), color='k', ls='--')

ax.annotate('Mínimo', xy=(x1,y1), fontsize=14, xytext=(x2,y2), arrowprops=dict(facecolor='k'))

plt.show()
```

### Aula 03

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
dias_da_semana = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']

temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[dias_da_semana]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indices = range(len(temperatura_por_dia_da_semana))

ax.bar(indices, temperatura_por_dia_da_semana)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
ax.set_xlabel('Dia da Semana', fontsize=12)
ax.set_ylabel('Temperatura média', fontsize=12)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indices = range(len(temperatura_por_dia_da_semana))

ax.bar(indices, temperatura_por_dia_da_semana)
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
ax.set_xlabel('Dia da Semana', fontsize=12)
ax.set_ylabel('Temperatura média', fontsize=12)

ax.set_xticks(indices)
ax.set_xticklabels(dias_da_semana, fontsize=10)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

indices = range(len(temperatura_por_dia_da_semana))
colors = ['b', 'r', 'g', 'b', 'orange', 'purple', 'yellow']


ax.bar(indices, temperatura_por_dia_da_semana, color=colors, edgecolor='k')
ax.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
ax.set_xlabel('Dia da Semana', fontsize=12)
ax.set_ylabel('Temperatura média', fontsize=12)

ax.set_xticks(indices)
ax.set_xticklabels(dias_da_semana, fontsize=10)

plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0,0,1,1])

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)
ax.set_title('Temperatura por dia da semana', fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
ax = fig.add_axes([0,0,1,1])

fim_de_semana = [0.1, 0, 0, 0, 0, 0, 0.1]

ax.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, 
       autopct='%.1f%%', explode=fim_de_semana, shadow=True)
ax.set_title('Temperatura por dia da semana', fontsize=16, pad=10)

plt.show()
```
