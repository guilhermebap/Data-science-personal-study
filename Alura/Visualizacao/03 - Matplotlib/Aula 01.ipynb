# Curso - Data Visualization: criação de gráficos com o Matplotlib

## Aula 01 

### Fazendo as primeiras visualizações

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
df = pd.read_csv('dados/monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.plot(df['data'], df['temperatura'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo')
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'])

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'])
ax.set_title('Temperatura no momento')
ax.set_xlabel('Data')
ax.set_ylabel('Temperatura')

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='green')
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate')
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)

plt.show()
```

### Aula 02

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate', lw = 4)
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate', lw = 0.4)
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)

plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate')
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)
ax.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))

plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate', marker='o')
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)
ax.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
plt.show()
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(df['data'], df['temperatura'], color='chocolate')
ax.set_title('Temperatura no momento', fontsize=18)
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Temperatura', fontsize=14)
ax.legend(['Temperatura'], loc='lower right', fontsize=14)
ax.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
ax.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1, 0.3, 0.4, 0.3])

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1, 0.3, 0.4, 0.3])

ax1.grid(True)
ax1.plot(df['data'], df['temperatura'])
ax1.set_title('Temperatura', fontsize=18, pad=20)

plt.show()
```
