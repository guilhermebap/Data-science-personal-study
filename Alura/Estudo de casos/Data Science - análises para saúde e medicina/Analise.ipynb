```python
def lastest_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:, -1]
```

```python
lastest_by_country(confirmed).head()
```

```python
informations = [lastest_by_country(confirmed), lastest_by_country(deaths), lastest_by_country(recovered)]
```

```python
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined
```

```python
sum_up = combined.sum()
letality_rate = sum_up['deaths'] / sum_up['confirmed']
print(f'A taxa de letalidade no conjunto de dados é: {letality_rate:0.2%}')
```

```python
letality_rate_1 = combined['deaths'] / combined['confirmed'] * 100
combined['letality_rate_1'] = letality_rate_1
combined['letality_rate_2'] = combined['deaths'] / (combined['deaths'] + combined['recovered']) * 100
combined.head()
```

```python
def letality_rate_at(data):
    confirmed_at = confirmed.groupby('Country/Region').sum().loc[:, data]
    deaths_at = deaths.groupby('Country/Region').sum().loc[:, data]
    letality_rate = deaths_at.sum() / confirmed_at.sum()
    return letality_rate
```

```python
datas = ['1/30/20', '2/28/20', '3/30/20', '4/30/20', '5/30/20', '6/30/20', '7/30/20', '8/30/20',
         '9/30/20', '10/30/20', '11/30/20', '12/30/20', '1/30/21', '2/28/21', '3/30/21', '4/30/21',
         '5/30/21', confirmed.columns[-1]]
litality_rate_months = []
for i in datas:
    litality_rate_months.append(letality_rate_at(i))
litality_rate_months
```

```python
fig, ax = plt.subplots(figsize=(18,8))
sns.lineplot(x=datas, y=litality_rate_months, ax=ax)
ax.set_title('Evolução da taxa de letalidade ao longo do tempo', fontsize=18, pad=10)
ax.set_xlabel('Meses', fontsize=14)
ax.set_ylabel('taxa de letalidade', fontsize=14)
ax.tick_params(labelsize=12, labelrotation=45)
plt.show()
```

```python
def rename_location(location):
    if location == 'Russian Federation':
        return 'Russia'
    elif location == 'United States of America':
        return 'US'
    elif location == 'Iran (Islamic Republic of)':
        return 'Iran'
    elif location == 'Bolivia (Plurinational State of)':
        return 'Bolivia'
    elif location == 'Venezuela (Bolivarian Republic of)':
        return 'Venezuela'
    elif location == 'Republic of Korea':
        return 'Korea, South'
    elif location == 'China, Hong Kong SAR':
        return 'Hong Kong'
    return location
```

```python
un_population_total.index = un_population_total.index.map(rename_location)
combined_expanded = combined.join(un_population_total)
combined.join(un_population_total).sort_values('confirmed', ascending=False)[:30]
```

```python
combined_expanded['incidence_ratio'] = combined_expanded['confirmed'] / combined_expanded['PopTotal'] * 1e5
combined_expanded['mortality_rate'] = combined_expanded['deaths'] / combined_expanded['PopTotal'] * 1e5
combined_expanded.head()
```
